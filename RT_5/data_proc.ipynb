```python
import numpy as np 
from matplotlib import pyplot as plt
```

```python
# 1.3

U_out =np.array( [150  ,300,450 ,745,1460, 2700,3388 , 3700])
U_in = np.array([10 ,  20 ,  30 ,50, 100,200 , 300 , 500])

plt.scatter(U_in, U_out/U_in)
plt.plot(U_in, U_out/U_in, c = 'skyblue')
plt.grid()
plt.ylabel('K')
plt.xlabel(r'$U_{in}$, mV')

plt.savefig('pictures/gr1.png', dpi = 400)
plt.show()
```

```python
# 1.5 

f = np.array([400, 500, 700, 900, 1000, 1050, 1070, 1100, 1120, 1150])
U_out = np.array([7.2, 9.45, 17, 40, 94, 224, 289, 210, 137, 97.5])

plt.plot(f, [max(U_out/20)/np.sqrt(2) for i in range(len(f))], '--', c = 'darkorange')
plt.scatter(f, U_out/20)
plt.plot(f, U_out/20, c = 'skyblue')
plt.legend(['0.7 level'])
plt.grid()
plt.ylabel(r'$K$')
plt.xlabel(r'$f$, kGz')

plt.savefig('pictures/gr2.png', dpi = 400)
plt.show()
```
